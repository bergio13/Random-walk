# Random Walk

In mathematics, a random walk is a random process that describes a path that consists of a succession of random steps on some mathematical space.

An elementary example of a random walk is the random walk on the integer number line Z which starts at 0, and at each step moves +1 or âˆ’1 with equal probability. Other examples include the path traced by a molecule as it travels in a liquid or a gas (Brownian motion), the search path of a foraging animal, or the price of a fluctuating stock and the financial status of a gambler. Random walks have applications to engineering and many scientific fields including ecology, psychology, computer science, physics, chemistry, biology, economics, and sociology. The term random walk was first introduced by Karl Pearson in 1905.

## 1 Dimension Random Walk

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def RandomWalk1d(steps):
    x_coordinates, y_coordinates = [], []
    x_coord = 0
    y_coord = 0
    for i in range(steps):
        x = np.random.rand(1, 1)
        if x < 0.5:
            next_s = -1
        else:
            next_s = 1
        x_coord = i
        if i == 0:
            y_coord = 0
        else:
            y_coord += next_s
        x_coordinates.append(x_coord)
        y_coordinates.append(y_coord)
    return x_coordinates, y_coordinates
```

```python
x, y = RandomWalk1d(100000)
```

```python
fig, ax = plt.subplots(figsize=(10, 4), dpi=100)
ax.scatter(x, y, c='blue', alpha=0.25, s=0.1)
ax.plot(0, 0, c='black', marker='+');

plt.title('1D Random Walk');
```

### Simulate K 1d random walks

```python
x1, y1 = RandomWalk1d(1000)
x2, y2 = RandomWalk1d(1000)
x3, y3 = RandomWalk1d(1000)

fig, ax = plt.subplots(figsize=(10, 4), dpi=100)

ax.plot(x1, y1);
ax.plot(x2, y2);
ax.plot(x3, y3);

plt.title('K 1d Random Walks');
```

## 2 Dimensions Random Walk

```python
x, y, end_x, end_y = RandomWalk2d(1000)
```

```python
fig, ax = plt.subplots(figsize=(6, 6), dpi=100)
ax.scatter(x, y, c='purple', s=0.9)
ax.plot(x, y, c='blue', alpha=0.75)
ax.plot(0, 0, c='red', marker='+');
ax.plot(end_x, end_y, c='black', marker='o')

plt.title('2D Random Walk');
```

### Simulate K 2d random walks

```python
x1, y1, end_x1, end_y1 = RandomWalk2d(1000)
x2, y2, end_x2, end_y2 = RandomWalk2d(1000)
x3, y3, end_x3, end_y3 = RandomWalk2d(1000)

fig, ax = plt.subplots(figsize=(8, 8), dpi=100)

ax.scatter(x1, y1, c='purple', alpha=0.5, s=0.5)
ax.plot(x1, y1, c='blue', alpha=0.75, lw=0.5)
ax.plot(end_x1, end_y1, c='darkblue', marker='o')

ax.scatter(x2, y2, c='orange', alpha=0.5, s=0.5)
ax.plot(x2, y2, c='red', alpha=0.75, lw=0.5)
ax.plot(end_x2, end_y2, c='red', marker='o')

ax.scatter(x3, y3, c='lightblue', alpha=0.5, s=0.5)
ax.plot(x3, y3, c='green', alpha=0.75, lw=0.5)
ax.plot(0, 0, c='black', marker='D');
ax.plot(end_x3, end_y3, c='darkgreen', marker='o')

plt.title('K 2D Random Walks');
```

## 3 Dimensions Random Walk

```python
def RandomWalk3d(steps):
    x = np.zeros(steps)
    y = np.zeros(steps)
    z = np.zeros(steps)
    for i in range(1, steps):
        val = np.random.randint(1, 7)
        if val == 1:
            # Step right
            x[i] = x[i-1] + 1
            y[i] = y[i-1]
            z[i] = z[i-1]
        elif val == 2:
            # Step left
            x[i] = x[i-1] - 1
            y[i] = y[i-1]
            z[i] = z[i-1]
        elif val == 3:
            # Step up
            x[i] = x[i-1]
            y[i] = y[i-1] + 1
            z[i] = z[i-1]
        elif val == 4:
            # Step down
            x[i] = x[i-1]
            y[i] = y[i-1] - 1
            z[i] = z[i-1]
        elif val == 5:
            # Step in
            x[i] = x[i-1]
            y[i] = y[i-1]
            z[i] = z[i-1] - 1
        elif val == 6:
            # Step out
            x[i] = x[i-1]
            y[i] = y[i-1]
            z[i] = z[i-1] + 1
    end_x = x[i]
    end_y = y[i]
    end_z = z[i]
    return x, y, z, end_x, end_y, end_z
```

```python
x, y, z, end_x, end_y, end_z = RandomWalk3d(5000)
```

```python
fig = plt.figure(figsize=(8, 8), dpi=100)
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c='purple', alpha=0.5, s=0.5)
ax.plot(x, y, z, c='blue', alpha=0.75, lw=0.5)
ax.plot(0, 0, 0, c='red', marker='D');
ax.plot(end_x, end_y, end_z, c='black', marker='o')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.title('3D Random Walk');
```

### Simulate K 3d Random Walks

```python
x1, y1, z1, end_x1, end_y1, end_z1 = RandomWalk3d(1000)
x2, y2, z2, end_x2, end_y2, end_z2 = RandomWalk3d(1000)
x3, y3, z3, end_x3, end_y3, end_z3 = RandomWalk3d(1000)
x4, y4, z4, end_x4, end_y4, end_z4 = RandomWalk3d(1000)

fig = plt.figure(figsize=(8, 8), dpi=100)

ax = fig.add_subplot(111, projection='3d')

ax.scatter(x1, y1, z1, alpha=0.5, s=0.5)
ax.plot(x1, y1, z1, alpha=0.75, lw=0.5)
ax.plot(0, 0, 0, c='red', marker='D');
ax.plot(end_x1, end_y1, end_z1, c='black', marker='o')

ax.scatter(x2, y2, z2, alpha=0.5, s=0.5)
ax.plot(x2, y2, z2, alpha=0.75, lw=0.5)
ax.plot(end_x2, end_y2, end_z2, c='red', marker='o')

ax.scatter(x3, y3, z3, alpha=0.5, s=0.5)
ax.plot(x3, y3, z3, alpha=0.75, lw=0.5)
ax.plot(end_x3, end_y3, end_z3, c='darkgreen', marker='o')

ax.scatter(x4, y4, z4, alpha=0.5, s=0.5)
ax.plot(x4, y4, z4, alpha=0.75, lw=0.5)
ax.plot(end_x4, end_y4, end_z4, marker='o')



ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.title('K 3D Random Walks');
```
